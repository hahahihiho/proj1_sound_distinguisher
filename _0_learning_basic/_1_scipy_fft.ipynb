## fft 연습

```python
from scipy import fftpack
import numpy as np
import matplotlib.pyplot as plt
```

```python
second = 2
sample_rate = 44100
f = 10000
t = np.linspace(0,second,second*sample_rate)
cosine = np.cos(2*np.pi*f*t)
f_domain = fftpack.fftfreq(second*sample_rate)*sample_rate
plt.plot(f_domain,fftpack.fft(cosine))
```

```python
fftpack.fft(cosine).max()
```

```python
plt.plot(f_domain/sample_rate)
```

## fft plot

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack

# Number of samplepoints
N = 600
# sample spacing (간격)(1/sample_rate)
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = scipy.fftpack.fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)

fig, ax = plt.subplots()
ax.plot(xf, 2.0/N * np.abs(yf[:N//2]))
plt.show()
```
